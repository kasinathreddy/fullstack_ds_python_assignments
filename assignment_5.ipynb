```python
def lcm(a,b):
    start=0
    flag=True
    while flag:
        start+=a
        if start%b==0:
            flag=False
    return start
```

```python
lcm(2,3)
```

```python
def hcf(a,b):
    out=1
    for number in range(1,a+1):
        if a%number==0 and b%number==0:
            out=number
    return out
```

```python
hcf(60,75)
```

```python
def dec_conversion(dec_number,change_type='bin'):
    c_dict={'bin':2,'oct':8,'hex':16}
    out=''
    check=c_dict.get(change_type)
    if check is not None:
        temp=dec_number
        while temp>=check:
            out+=str(temp%check)
            temp=temp//check
        out+=str(temp%check)
        out=out[::-1]
        return int(out)
```

```python
dec_conversion(17,change_type='bin')
```

```python
dec_conversion(17,change_type='oct')
```

```python
dec_conversion(17,change_type='hex')
```

```python
def asci_val(character):
    return f'Ascii value of {character} is {ord(character)}'
```

```python
asci_val('a')
```

```python
def calc(number_1,number_2):
    operation=input()
    print(f'operation given is {operation}')
    if operation=='+':
        return number_1+number_2
    elif operation=='-':
        return number_1-number_2
    elif operation=='*':
        return number_1*number_2
    elif operation=='/':
        return number_1/number_2
```

```python
calc(1,2)
```

```python
calc(1,2)
```

```python
calc(1,2)
```

```python
calc(1,2)
```
