```python
def sum_of_array(array):
    output_sum=0
    for number in array:
        output_sum+=number
    return output_sum
```

```python
sum_of_array([2,2,2,2,2])
```

```python
def largest_element_in_array(array):
    max_ele=array[0]
    for element in array[1:]:
        if element>max_ele:
            max_ele=element
    return max_ele
```

```python
largest_element_in_array([5,1,7,8,20])
```

```python
def rotate_array(array,side,n_terms):
    if side=='left':
        return array[n_terms:]+array[:n_terms]
    elif side=='right':
        return array[len(array)-n_terms:][::-1]+array[:len(array)-n_terms]
```

```python
rotate_array([1,3,5,7,9,11],'left',4)
```

```python
def array_splitter(array):
    check=len(array)//2
    return array[check:]+array[:check]
```

```python
array_splitter([1,2,3,4])
```

```python
array_splitter([1,2,3,4,5])
```

```python
def monotonic_array(array):
    increment=True
    decrement=True
    temp1=array[0]
    for i in array[1:]:
        if i>temp1:
            temp1=i
        else:
            increment=False
            break
    if increment:
        return increment
    temp2=array[0]
    for i in array[1:]:
        if i<temp2:
            temp2=i
        else:
            decrement=False
            break
    if decrement:
        return decrement
    return decrement
```

```python
monotonic_array([1,2,3,5])
```

```python
monotonic_array([5,1,2,3,5])
```
