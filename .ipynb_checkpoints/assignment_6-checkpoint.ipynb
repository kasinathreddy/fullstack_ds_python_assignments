```python
def fibonacci(fib_level):
    if fib_level<=1:
        return fib_level
    else:
        return(fibonacci(fib_level-1) + fibonacci(fib_level-2))
def exec_fibonacci(fib_level):
    for i in range(fib_level):
         print(fibonacci(i))
```

```python
exec_fibonacci(5)
```

```python
def factorial(n):
    if n<=1:
        return 1
    else:
        return n*factorial(n-1)
```

```python
factorial(5)
```

```python
def calc_bmi(height,weight):
    #height in meters
    #weight in kgs
    return weight/height**2
```

```python
calc_bmi(1.8,60)
```

```python
import numpy as np
def nat_log(number):
    return np.log(number)
```

```python
nat_log(10)
```

```python
def cube_sum(n):
    out=0
    for number in range(1,n+1):
        out+=number**3
    return out
```

```python
cube_sum(3)
```
