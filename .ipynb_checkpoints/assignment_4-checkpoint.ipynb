```python
def factorial(number):
    out=1
    if number>0:
        for n in range(1,number+1):
            out*=n
    return out
```

```python
factorial(5)
```

```python
factorial(1)
```

```python
factorial(0)
```

2.	Write a Python Program to Display the multiplication Table?

```python
def mult_table(table_number,table_depth):
    for multiplicator in range(1,table_depth+1):
        print(f"{table_number} X {multiplicator} = {table_number*multiplicator}")
```

```python
mult_table(2,9)
```

```python
mult_table(2,20)
```

```python
def fibonnaci(fib_level):
    a=0
    b=1
    for _ in range(fib_level):
        print(a)
        temp=a
        a=b
        b+=temp
        
```

```python
fibonnaci(10)
```

```python
def armstrong_number(main_number):
    check=0
    number=str(main_number)
    pow_fact=len(number)
    for n in number:
        check+=int(n)**pow_fact
    return main_number==check
```

```python
armstrong_number(153)
```

```python
armstrong_number(21)
```

```python
def armstrong_in_range(start,stop):
    for number in range(start,stop+1):
        if armstrong_number(number):
            print(number)
```

```python
armstrong_in_range(1,200)
```

```python
armstrong_in_range(100,1000)
```

```python
def sum_of_natural_numbers(*numbers):
    sum_output=0
    for number in numbers:
        if type(number)==int and number>0:
            sum_output+=number
    return sum_output
```

```python
sum_of_natural_numbers(1,2,'5',2.0,3)
```
