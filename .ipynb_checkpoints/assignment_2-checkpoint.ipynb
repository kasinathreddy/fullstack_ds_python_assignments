```python
def km_to_miles(input_kms):
    return input_kms/1.609
```

```python
km_to_miles(1)
```

```python
def celsius_to_fahrenheit(celsius_reading):
    return ((9*celsius_reading)/5)+32
```

```python
celsius_to_fahrenheit(25)
```

```python
#eqn consists of ax**2+bx+c
def quad_func(a,b,c):
    x1=(-b+(b**2-(4*a*c)))/2*a
    x2=(-b-(b**2-(4*a*c)))/2*a
    return x1,x2
```

```python
quad_func(1,5,6)
```

```python
def swap_two_vars(var_1,var_2):
    var_1,var_2=var_2,var_1
    return var_1,var_2
```

```python
var1=1
var2=2
var1,var2=swap_two_vars(var1,var2)
print(var1)
print(var2)
```
