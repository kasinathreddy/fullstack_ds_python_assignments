```python
def add(*args):
    output_sum=0
    for ele in args:
        output_sum+=ele
    return output_sum
def divide(divisor,dividend):
    return divisor/dividend
```

```python
add(10,40,50)
```

```python
divide(10,5)
```

```python
def area_of_triangle(base,height):
    return 1/2*base*height
```

```python
area_of_triangle(3,4)
```

```python
def swap_two_variables(var_1,var_2):
    var_1,var_2=var_2,var_1
    return var_1,var_2
```

```python
var_1=1
var_2=2
var_1,var_2=swap_two_variables(var_1,var_2)
print(var_1)
print(var_2)
```

```python
from random import randint
def return_random_number(starting_number,ending_number):
    return randint(starting_number,ending_number)
```
